```python
import numpy as np
import matplotlib.pyplot as plt


models = ["LR", "KNN", "XGBOOST", "RF", "SVM"]
f1_base =  [0.85,   0.86,   0.8982, 0.8874, 0.8921]
f1_hog  =  [0.87,   0.88,   0.8250, 0.8663, 0.8905]
f1_sift =  [0.6491,    0.6115,    0.6548, 0.6344, 0.6474]
f1_pca  =  [0.85,   0.87,   0.8786, 0.8656, 0.8971]

x = np.arange(len(models))  
width = 0.2                 

plt.figure(figsize=(8, 5))

# Plot
plt.bar(x - 1.5*width, f1_base, width, label="raw (base)")
plt.bar(x - 0.5*width, f1_hog,  width, label="HOG")
plt.bar(x + 0.5*width, f1_sift, width, label="SIFT")
plt.bar(x + 1.5*width, f1_pca,  width, label="PCA")

# labeling
plt.xticks(x, models)
plt.ylabel("F1 score")
plt.title("F1-scores for ML Models")
plt.ylim([0, 1.05]) 
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.subplots_adjust(bottom=0.15)


plt.legend(
    loc='upper center',    
    bbox_to_anchor=(0.5, -0.15),  # 
    fancybox=True, 
    shadow=True, 
    ncol=4                  
)

plt.show()
```
